```python
results = []

for title, content in notes.items():

    full_text = title + " " + content

    score = 0

for token in tokens:

    token_score = fuzz.partial_ratio(

        token,

        full_text.lower()

    )

    score += token_score

    results.append(
        (title, score)
    )


results.sort(
    key=lambda x: x[1],
    reverse=True
)


for result in results[:3]:

    print(result)
```
