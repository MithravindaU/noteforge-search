```python
notes = {

    "Sorting Algorithms":
    "Bubble sort selection sort merge sort quick sort",

    "Machine Learning":
    "Supervised learning unsupervised learning regression",

    "Database Systems":
    "SQL normalization joins primary key"

}
```

```python
query = input(
    "Enter topic: "
)
```

```python
from rapidfuzz import fuzz
```

```python
best_match = ""

highest_score = 0
```

```python
for title, content in notes.items():

    full_text = title + " " + content

    score = fuzz.partial_ratio(

        query.lower(),

        full_text.lower()

    )

    if score > highest_score:

        highest_score = score

        best_match = title
```

```python
if score > highest_score:

    highest_score = score

    best_match = title
```

```python
print(
    "Best Match:",
    best_match
)

print(
    "Similarity Score:",
    highest_score
)
```
